```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
fmri=sns.load_dataset("fmri")
fmri.head()
```

```python
fmri.shape
```

```python
#line plot
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri)

plt.show()
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event')
plt.show()
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event',style='event')
plt.show()
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event',style='event',markers=True)
plt.show()
```

```python
#histogram / distplot
```

```python
diamond=pd.read_csv('diamonds.csv')
diamond.head()
```

```python
sns.distplot(diamond['price'])
plt.show()
```

```python
sns.distplot(diamond['price'],hist=False)
plt.show()
```

```python
sns.distplot(diamond['price'],color='green')
plt.show()
```

```python
sns.distplot(diamond['price'],color='red',bins=10,kde=False)
plt.show()
```

```python
sns.distplot(diamond['price'],color='pink',vertical=True)
plt.show()
```

```python
#box plot
```

```python
churn =pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
churn.head()
```

```python
sns.boxplot(x='Churn',y='tenure',data=churn)
plt.show()
```

```python
sns.boxplot(x='InternetService',y='MonthlyCharges',data=churn)
plt.show()
```

```python
sns.boxplot(x='Contract',y='tenure',data=churn,palette='Set1')
plt.show()
```

```python
sns.boxplot(x='Contract',y='tenure',data=churn,palette='Set2')
plt.show()
```

```python
sns.boxplot(x='Contract',y='tenure',data=churn,palette='Set1',linewidth=3)
plt.show()
```

```python
sns.boxplot(x='Contract',y='tenure',data=churn,order=['Two year','Month-to-month','One year'])
plt.show()
```

```python
sns.boxplot(x='Contract',y='tenure',data=churn,hue='PaymentMethod')
plt.show()
```
