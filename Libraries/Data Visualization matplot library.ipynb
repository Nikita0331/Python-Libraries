```python
#Data Visualization with Python : Matplot lib
```

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
#line plot
```

```python
x=np.arange(1,11)
x
```

```python
y=2*x
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(y,x)
plt.show()
```

```python
plt.plot(x,y)
plt.title("LINE PLOT")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
plt.plot(x,y,color='g',linestyle=':',linewidth=5)
plt.show()
```

```python
plt.plot(x,y,color='r',linestyle='-',linewidth=2)
plt.title("LINE PLOT")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
x=np.arange(1,11)
y1=2*x
y2=3*x
```

```python
plt.plot(x,y1,color='r',linestyle='-.',linewidth=2)
plt.plot(x,y2,color='g',linestyle=':',linewidth=3)
plt.title("LINE PLOT")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
x=np.arange(1,11)
y1=2*x
y2=3*x
plt.subplot(1,2,1)
plt.plot(x,y1,color='r',linestyle='-.',linewidth=2)
plt.title("LINE PLOT")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.subplot(1,2,2)
plt.plot(x,y2,color='g',linestyle=':',linewidth=3)
plt.title("LINE PLOT")
plt.xlabel("x-axis")
# plt.ylabel("y-axis")
plt.grid(True)

plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1,color='r',linestyle='-.',linewidth=2)
plt.title("LINE PLOT")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.subplot(2,1,2)
plt.plot(x,y2,color='g',linestyle=':',linewidth=3)
plt.title("LINE PLOT")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)

plt.show()
```

```python
#bar plot
```

```python
student={'bob':87,'nik':56,'mik':66}
```

```python
names=list(student.keys())
marks=list(student.values())
```

```python
plt.bar(names,marks)
plt.show()
```

```python
plt.bar(names,marks,color='b')
plt.title('Distribution of student marks')
plt.xlabel('names')
plt.ylabel('marks')
plt.grid(True)
plt.show()
```

```python
plt.barh(names,marks,color='g')
plt.title('Distribution of student marks')
plt.ylabel('names')
plt.xlabel('marks')
plt.grid(True)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,80,90]
a=[1,3,5,7,2,4,8,9,0]
b=[9,3,5,7,1,6,8,4,2]
plt.scatter(x,a)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,80,90]
a=[1,3,5,7,2,4,8,9,0]
b=[9,4,7,8,1,6,3,4,2]
plt.scatter(x,a,marker='*',c='g',s=100)
plt.scatter(x,b,marker='.',c='r',s=200)
# plt.grid(True)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(x,a,marker='*',c='g',s=100)
plt.subplot(1,2,2)
plt.scatter(x,b,marker='.',c='r',s=200)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.scatter(x,a,marker='*',c='g',s=100)
plt.subplot(2,1,2)
plt.scatter(x,b,marker='.',c='r',s=200)
plt.show()
```

```python
#pie chart
```

```python
marks=[90,80,70,60]
students=[10,15,30,20]
```

```python
plt.pie(students,labels=marks)
plt.title('no. of students and marks')
plt.show()
```

```python
plt.pie(students,labels=marks,autopct='%0.1f%%',colors=['green','grey','black','blue'])
plt.title('no. of students and marks')
plt.show()
```

```python
#dough nut chart
```

```python
marks=[90,80,70,60]
students=[10,15,30,20]
```

```python
plt.pie(students,labels=marks,radius=2)
plt.pie([1,1],colors=['black','w'],radius=1)
# plt.title('no. of students and marks')
plt.show()
```
